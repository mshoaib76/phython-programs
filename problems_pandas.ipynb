```python
import pandas as pd 
import numpy as np
```

```python

dict={
     'animal':['cat','cat','snake','dog','dog','cat','snake','dog','lion','lion'],
       'age':[2.5,3,0.5,np.nan,5,2,4.5,np.nan,7,3],
       'visits':[1,3,2,3,2,3,1,1,2,4],
       'priority':['yes','yes','no','yes','no','no','no','yes','yes','no'],   
}
labels=['a','b,','c','d','e','f','g','j','i','j']
df=pd.DataFrame(dict,index=labels)
df
      
```

```python
df.isnull().sum()
```

```python
df.sort_values(by=['age','visits'],ascending=[False, True])
```

replace the priority column yes into True and no into False

```python
df['priority']=df['priority'].replace({'yes':True,'no':False})
df
```

replace the animal column snake into phython

```python
df['animal'].replace({'snake':'phython'},inplace=True)
df
```

for each animal type and each number of visits ,find the mean age ,in other words each row in number of animal each colums is number of visits and the values are the mean age(hints:use povit_table)

```python
df.pivot_table(values='age',index='animal',columns='visits',aggfunc='mean')
```
